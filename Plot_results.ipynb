```python
import numpy as np
from matplotlib import pyplot as plt
from pylab import genfromtxt 
from matplotlib import style
```

```python
data = genfromtxt("build/output_result_LR_DS2.txt")

fig= plt.figure(num=None, figsize=(10, 8), dpi=80, facecolor='w', edgecolor='k')

plt.title("UKF Lesar and Radar Dataset 2")
plt.scatter(data[:,4], data[:,5], label = "Laser measurementes ",c='red',s=2)
# plt.plot(data[:,4], data[:,5], label = "Laser measurementes ");
plt.plot(data[:,6], data[:,7], label = "Ground Truth",c='green');
plt.plot(data[:,0], data[:,1], label = "Estimated");
plt.legend();
fig.savefig('./UKF_LR_DS2.png')
plt.show();
```

```python
data = genfromtxt("build/NIS_R_DS1.txt",skip_header=1)
data2 = genfromtxt("build/NIS_R_DS2.txt",skip_header=1)
style.use('ggplot')
fig = plt.figure(figsize=(14, 3))
fig.suptitle('NIS Radar', fontsize=20, fontweight='bold')

ax = fig.add_subplot(121)
fig.subplots_adjust(top=0.85)
ax.set_title('Dataset 1', fontsize=18, fontweight='bold')
ax.set_xlabel('time')
ax.set_ylabel('chi2')
ax.plot(data)
ax.plot((0, 500), (7.8, 7.8), 'k-', linewidth=2, label = '7.8')
ax.legend(loc=0)
ax.axis([0, 500 , 0, 35])




ax = fig.add_subplot(122)
fig.subplots_adjust(top=0.85)
ax.set_title('Dataset 2', fontsize=18, fontweight='bold')
ax.set_xlabel('time')
ax.set_ylabel('chi2')
ax.plot(data2)
ax.plot((0, 500), (7.8, 7.8), 'k-', linewidth=2, label = '7.8')
ax.legend(loc=0)
ax.axis([0, 500 , 0, 35])
plt.savefig('NIS_R.png')
plt.show()

```
